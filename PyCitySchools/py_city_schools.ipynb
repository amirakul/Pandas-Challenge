```python
import pandas as pd
```

```python
# Make a reference to the books.csv file path
csv_path1 = "Resources/students_complete.csv"
csv_path2 = "Resources/schools_complete.csv"
```

```python
# Import the students_complete.csv file as a DataFrame
df1 = pd.read_csv(csv_path1, encoding="utf-8")
```

```python
# Import the schools.csv file as a DataFrame
df2 = pd.read_csv(csv_path2, encoding="utf-8")
```

# District Summary


```python
#Merging two csv files
merge_df = pd.merge(df2, df1, on="school_name")
merge_df
```

```python
#Looking for missing values
merge_df.count()
```

```python
#Calculating total number of schools
count_of_schools= len(merge_df["School ID"].unique())
count_of_schools
```

```python
#Calculating total number of students
merge_df["student_name"].value_counts()
```

```python
#Calculating total number of students
count_of_students=len(merge_df)
count_of_students
```

```python
#Calculating total budget
total_budget=df2["budget"].sum()
total_budget
```

```python
#Finding average math score
aver_math_score=merge_df["math_score"].mean()
aver_math_score
```

```python
#Finding average reading score
aver_reading_score=merge_df["reading_score"].mean()
aver_reading_score
```

```python
#Passing Math (The percentage of students that passed math.)
amount_passed_math = len(merge_df.loc[merge_df["math_score"] >=70])
amount_passed_math
```

```python
perc_of_students_passing_math=(amount_passed_math/count_of_students)*100
perc_of_students_passing_math
```

```python
#Passing Reading (The percentage of students that passed reading.)
amount_passed_reading = len(merge_df.loc[merge_df["reading_score"] >=70])
amount_passed_reading
```

```python
perc_of_students_passing_reading=(amount_passed_reading/count_of_students)*100
perc_of_students_passing_reading
```

```python
#Overall Passing (The percentage of students that passed math and reading.)
amount_passed_math_and_reading=merge_df.loc[(merge_df["math_score"] >=70) & (
    merge_df["reading_score"] >=70), :]
len(amount_passed_math_and_reading)
                        
```

```python
perc_of_students_passing_both=(len(amount_passed_math_and_reading)/count_of_students)*100
perc_of_students_passing_both
```

```python
# Create a new table consolodating above calculations
district_summary_0 = pd.DataFrame({"Total Number of Schools": [count_of_schools],
                                   "Total Number of Students": [count_of_students],
                                   "Total Budget": [total_budget],
                                   "Avg. Math Score": [aver_math_score],
                                   "Avg. Reading Score": [aver_reading_score],
                                   "% Passing Math": [perc_of_students_passing_math],
                                   "% Passing Reading":[perc_of_students_passing_reading],
                                   "% Passing Both": [perc_of_students_passing_both]
                                   })
district_summary = district_summary_0.round(2)
#district_summary['Total Budget'] = ['${:,.2f}'.format(x) for x in district_summary['Total Budget']]

district_summary['Total Budget'] = district_summary['Total Budget'].map("${:,.2f}".format)
district_summary
```

# School Summary
Create an overview table that summarizes key metrics about each school, including:
School Name
School Type
Total Students
Total School Budget
Per Student Budget
Average Math Score
Average Reading Score
% Passing Math (The percentage of students that passed math.)
% Passing Reading (The percentage of students that passed reading.)
% Overall Passing (The percentage of students that passed math and reading.)


```python
#Finding school budget for each school
#schools=merge_df.groupby(["school_name"])
d1 = df2.loc[:,[ "school_name",'type', 'size', 'budget']]

schools = pd.DataFrame(d1)

schools.rename(columns={'type': 'School Type',
                        'school_name':'School Name',
                        'size': 'Total Students',
                        'budget': 'Total School Budget', 
                        }, inplace=True)

schools['Per Student Budget'] = schools['Total School Budget']/schools['Total Students']

schools.head()
```

```python
#Calculating and listing average test score for each school
d2 = merge_df.groupby(['school_name'])
score = pd.DataFrame(round(d2.mean(),2))   
score1 = score.loc[:,['reading_score','math_score']]


score1=score1.reset_index()
score1.rename(columns={'school_name':'School Name',
                      'reading_score':'Aver. Reading Score',
                      'math_score': 'Aver. Math Score',
                     }, inplace=True)
score1
```

```python
add= pd.merge(schools,score1,how='left', on='School Name')
add
```

```python
#Counting how many students passed reading for each school

count_passing = merge_df.loc[:,['school_name','reading_score']]
count_passing['reading_score']=count_passing.loc[count_passing['reading_score']>=70]
count_passing=count_passing.groupby('school_name')
count_passing=pd.DataFrame(count_passing.count())
count_passing
```

```python
#Counting number of students for each school
count_student = merge_df.loc[:,['school_name','size']]
count_student=count_student.groupby('school_name')
count_student=pd.DataFrame(count_student.count())
count_student
```

```python
#Counting how many students passed math for each school

count_passing_math = merge_df.loc[:,['school_name','math_score']]
count_passing_math['math_score']=count_passing_math.loc[count_passing_math['math_score']>=70]
count_passing_math=count_passing_math.groupby('school_name')
count_passing_math=pd.DataFrame(count_passing_math.count())
count_passing_math
```

```python
#Joining counts for math, reading>=70 and total students' count
joined_counts=pd.merge(count_passing_math, count_passing,
how='left', on='school_name')
joined_counts1=pd.merge(joined_counts,count_student, how='left',on='school_name')
joined_counts1=joined_counts1.reset_index()
joined_counts1.rename(columns={'school_name':'School Name',
                               'math_score':'Total Count Passing Math',
                               'reading_score':'Total Count Passing Reading',
                               'size':'Total Students'},inplace=True)
joined_counts1
```

```python
#Figuring out counts of students that passed both math and reading
count_both=merge_df.loc[:,['school_name','reading_score','math_score']]
count_both2=count_both[(count_both['reading_score']>=70) & (count_both['math_score']>=70)]
count_both2=count_both2.groupby('school_name')
count_both2=pd.DataFrame(count_both2.count())

count_both3=count_both2.drop(['reading_score'], axis=1)
count_both3=count_both3.reset_index()
count_both3.rename(columns={'school_name':'School Name',
                               'math_score':'Overall Count',
                               },inplace=True)

count_both3
```

```python
#Merging joint counts and overall counts
joined_df=pd.merge(count_both3,joined_counts1, how='left',on='School Name')
joined_df
```

```python
joined_df["% Overall Passing"]=round(joined_df['Overall Count']/joined_df['Total Students']*100,2)
joined_df["% Passing Math"]=round(joined_df['Total Count Passing Math']/joined_df['Total Students']*100,2)
joined_df["% Passing Reading"]=round(joined_df['Total Count Passing Reading']/joined_df['Total Students']*100,2)
joined_df
```

```python
#Joining school summary and tests summary tables together
school_summary_df=pd.merge(add,joined_df, how='left',on=['School Name','Total Students'])
school_summary_df=school_summary_df.drop(['Overall Count', 'Total Count Passing Math','Total Count Passing Reading'], axis=1)
school_summary_df=school_summary_df[['School Name',
                                     'School Type',
                                     'Total Students',
                                     'Total School Budget',
                                     'Per Student Budget',
                                     'Aver. Math Score',
                                     'Aver. Reading Score',
                                     '% Passing Math',
                                     '% Passing Reading',
                                      '% Overall Passing']]
school_summary_df
```

# Scores by School Spending
Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
Average Math Score
Average Reading Score
% Passing Math (The percentage of students that passed math.)
% Passing Reading (The percentage of students that passed reading.)

```python
#Figuring out max and min school budget to create bins
print(schools["Per Student Budget"].max())
print(schools["Per Student Budget"].min())
```

```python
# Creating bins in which to place values based on per student budget(spending)
bins = [0, 585, 630, 645, 675]

# Create labels for these bins
schools2=school_summary_df.copy()
group_labels = ["< 585", "585 - 630", "630 - 645", "645 - 675"]
schools2['Spending Ranges Per Student']=pd.cut(schools["Per Student Budget"], bins, labels=group_labels)
schools2
```

```python
summary_by_spending=schools2.groupby(by="Spending Ranges Per Student").mean()
summary_by_spending=summary_by_spending.drop(['Total Students','Total School Budget','Per Student Budget'],axis=1)
summary_by_spending.round({'Aver. Math Score':2,'Aver. Reading Score':2,'% Passing Math':2,'% Passing Reading':2,"% Overall Passing":2})
```
